```python
from pyspark.sql import SparkSession, functions as F
from pyspark.sql.types import *
```

```python
spark = SparkSession.builder \
.appName("Create Filter") \
.master("local[2]") \
.getOrCreate()
```

```python
df = spark.read \
.option("header",True) \
.option("inferSchema", True) \
.option("compression", "gzip") \
.csv("file:///home/train/datasets/Hotel_Reviews.csv.gz")
```

```python
df2 = df.withColumn("Tags",
                    F.split(F.col("Tags"), ",")
                    .cast(ArrayType(StringType()))) \
.withColumn("Review_Date", F.to_date(F.col("Review_Date"), "M/d/yyyy"))
```

```python
df2.limit(2).toPandas()
```

```python
df2.filter("Average_Score > 6.0 ").limit(2).toPandas()
```

```python
df2.filter("Average_Score > 6.0 ").count()
```

```python
df2.filter("Reviewer_Nationality LIKE '%United Kingdom%' ").limit(3).toPandas()
```

```python
df2.selectExpr("Hotel_Name","TRIM(Reviewer_Nationality) AS Reviewer_Nationality") \
.filter("Reviewer_Nationality == 'United Kingdom' ").limit(3).toPandas()
```

```python
df2.filter(F.col("Hotel_Address").rlike("Milan")).limit(3).toPandas()
#rlike milan ge√ßen varsa getir
```

```python
df2.select("Hotel_Address").filter(F.col("Hotel_Address").rlike("Milan")).show(n=3, truncate=False)
```

```python
df2.filter("Reviewer_Nationality LIKE '%United Kingdom%' AND Average_Score > 8.0").limit(3).toPandas()
```
